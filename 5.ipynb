```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X=np.arange(10)
print(X)
Y=(X-5)**2
print(Y)
```

```python
plt.style.use("seaborn")
plt.plot(X,Y)
plt.show()
```

```python
x=0
lr=0.1
e=[]
plt.plot(X,Y)
for i in range(50):
    grad=2*(x-5)
    x=x-lr*grad
    y=(x-5)**2
    e.append(y)
    plt.scatter(x,y)
    print(x)
```

```python
plt.plot(e)
plt.show()
```
