```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn import metrics
```

```python
data=pd.read_excel(r"dataset.csv.xlsx",names=['Message','Label'])
print(data)
```

```python
data['Labelnum']=data.Label.map({'pos':1,'neg':0})
x=data.Message
y=data.Labelnum
print(x)
print(y)
```

```python
vectorizer=TfidfVectorizer()
data=vectorizer.fit_transform(x)
df=pd.DataFrame(data.toarray(), columns=vectorizer.get_feature_names())
df.head()
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(data,y,test_size=0.3,random_state=3)
print("Total number of training data ",ytrain.shape)
print("totoal number of test data ",ytest.shape)
```

```python
clf=MultinomialNB().fit(xtrain,ytrain)
predicted=clf.predict(xtest)
print("accuracy: ",metrics.accuracy_score(ytest,predicted))
print("confusion matrix: ",metrics.confusion_matrix(ytest,predicted))
print("classification report: ",metrics.classification_report(ytest,predicted))
print("precision score: ",metrics.precision_score(ytest,predicted))
print("recall score: ",metrics.recall_score(ytest,predicted))
```
